```python
#42. Add Elements Conditionally
#Write a NumPy program to add elements to a matrix. If an element in the matrix is 0, we will not add the element below this element
a=np.array([[1,1,0,2],
           [0,3,0,3],
             [1,0,4,4],
            ])
sum_no=0
row,col=a.shape
for i in range(row):
    for j in range(col):
        if a[i][j]==0 and i < row-1:
            a[i+1][j]=0
        sum_no=sum_no + a[i][j] 
print(sum_no)
```
